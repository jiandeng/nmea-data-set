### Import

```python
from altair import *
import pandas as pd
import numpy as np
import sys

sys.path.append('..')
import nmea_util as nu

from plotly.offline import init_notebook_mode
from plotly.offline import iplot

init_notebook_mode(connected=True)
```

### Setting

```python
file_name = '../data/new.firmware.dfn.roof.cold.start/1.GPS-20170526T204415.txt'
```

### Query

```python
print('Coverting records, might take a while...')
r = nu.nmea_to_dataframe(file_name)
print('done!')
```

### Filter

```python
r.tail()
```

### Visual

```python
data = [{
            'x': r[r.fixseq==seq]['fixtime'],
            'y': r[r.fixseq==seq]['vstar'],
            'name': 'Stars in view({})'.format(seq), 'mode': 'markers',
        } for seq in r.fixseq.unique()]
data.extend([{
            'x': r[r.fixseq==seq]['fixtime'],
            'y': r[r.fixseq==seq]['ustar'],
            'name': 'Used Stars({})'.format(seq), 'mode': 'markers',
        } for seq in r.fixseq.unique()])

iplot(
    {
        'data': data,
        'layout': {
            'xaxis': {'title': 'Time', 'rangeslider': {}},
            'yaxis': {'title': "Used stars vs Stars in view"}
            
        }
    }
)
```
